```python
import pandas as pd
import numpy as np

# 예시 DataFrame 생성
df = pd.DataFrame({
    'age': [12, 13, 14, 16],
    'weight': [41.1, 34.5, np.nan, 90.1],
    'height': ['1.72', '1.74', '1.91', '1.54']
})

# 1. 'weight' 열의 2번째 인덱스에 NaN 값 넣기
df.loc[2, 'weight'] = np.nan

# 2. 1번째 행의 모든 값을 NaN으로 설정
df.loc[1, :] = np.nan
print(df)
# 3. NaN이 있는 행 제거 (기본값은 axis=0, how='any')
df_without_na_any = df.dropna(how='any', axis=0)
print(df_without_na_any)


# 4. NaN이 모두 있는 행만 제거 (how='all')
df_without_na_all = df.dropna(how='all', axis=0)
print(df_without_na_all)


# 5. 최소한 3개의 유효한 값이 있는 행만 남기기
df_without_na_thresh = df.dropna(thresh=3, axis=0)
print(df_without_na_thresh)

# 6. NaN이 하나라도 있는 행 제거 (axis=0)
df_without_na_any_again = df.dropna(how='any', axis=0)
print(df_without_na_any_again)

# 7. NaN이 없으면 해당 행만 남기기 (thresh=df.shape[1]로 설정)
df_without_all_na = df.dropna(thresh=df.shape[1], axis=0)
print(df_without_all_na)

# 8. NaN이 있는 열 제거 (axis=1)
df_without_na_cols = df.dropna(axis=1)
print(df_without_na_cols)

# 9. NaN이 하나라도 있는 열 제거 (axis=1)
df_without_na_cols_any = df.dropna(how='any', axis=1)
print(df_without_na_cols_any)

# 10. NaN이 모두 있는 열 제거 (axis=1)
df_without_na_cols_all = df.dropna(how='all', axis=1)
print(df_without_na_cols_all)

# 11. NaN이 3개 이상이 아닌 열을 제거 (thresh=3)
df.dropna(axis=1, thresh=3, inplace=True)
print(df)
```

```python
import pandas as pd

# 예시 데이터
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)

# 1. 'LB' 포지션, 나이가 25 이하인 선수들
filtered_players_1 = players[
    (players.position == 'LB') &
    (players.age <= 25)
]
print(filtered_players_1)

# 2. 'LB' 포지션, 나이가 25 이하, market_value가 10 이상인 선수들
filtered_players_2 = players[
    (players.position == 'LB') &
    (players.age <= 25) &
    (players.market_value >= 10)
]
print(filtered_players_2)

# 3. 'LB' 포지션, 나이가 25 이하, market_value가 10 이상, 'Tottenham'과 'Arsenal' 클럽이 아닌 선수들
filtered_players_3 = players[
    (players.position == 'LB') &
    (players.age <= 25) &
    (players.market_value >= 10) &
    ~(players.club.isin(['Tottenham', 'Arsenal']))
]
print(filtered_players_3)
```

```python
import pandas as pd

# 예시 데이터
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)

# 1. 중복된 행 확인
duplicates = players.duplicated()
print(duplicates)

# 2. 중복된 행만 선택
duplicates_rows = players[players.duplicated()]
print(duplicates_rows)

# 3. 특정 열에 대한 중복 행 확인
duplicate_specific_columns = players.loc[players.duplicated(subset=['club', 'age', 'position', 'market_value'])]
print(duplicate_specific_columns)

# 4. 중복된 행의 개수
duplicate_count = players.loc[players.duplicated(subset=['club', 'age', 'position', 'market_value'])].shape
print(duplicate_count)

# 5. 마지막 중복 행만 선택
last_duplicates = players.loc[players.duplicated(subset=['club', 'age', 'position', 'market_value'], keep='last')]
print(last_duplicates)

# 6. 모든 중복 행 선택
all_duplicates = players.loc[players.duplicated(subset=['club', 'age', 'position', 'market_value'], keep=False)]
print(all_duplicates)

# 7. 중복된 행 삭제 (첫 번째 유지)
players_no_duplicates = players.drop_duplicates(keep='first')
print(players_no_duplicates)

# 8. 특정 행 삭제 (19번 인덱스)
players_dropped_row = players.drop(labels=19, axis=0)
print(players_dropped_row)

# 9. 여러 행 삭제
players_dropped_rows = players.drop(index=[19, 20, 21, 231, 10])
print(players_dropped_rows)

# 10. 특정 열 삭제
players_dropped_columns = players.drop(labels=['age', 'market_value'], axis='columns')
print(players_dropped_columns)

# 11. 여러 열 삭제
players_dropped_columns_2 = players.drop(columns=['age', 'market_value', 'name'])
print(players_dropped_columns_2)
```

```python
import pandas as pd
import numpy as np

# 예시 데이터
data_url = 'https://andybek.com/pandas-soccer'
players = pd.read_csv(data_url)

# 1. 'age'와 'position_cat' 열에 대해 평균값 구하기
agg_mean = players[["age", "position_cat"]].agg('mean')
print(agg_mean)

# 2. 'new_signing' 열에 대해 평균값 구하기
new_signing_mean = players.new_signing.mean()
print(new_signing_mean)

# 3. 'age'와 'position_cat' 열에 대해 최솟값 구하기
agg_min = players[["age", "position_cat"]].agg(np.min)
print(agg_min)

# 4. 숫자형 데이터의 최솟값 구하기
min_values = players.select_dtypes(np.number).agg(np.min)
print(min_values)

# 5. 숫자형 데이터에 대해 'min', 'max', 'mean'을 동시에 구하기
min_max_mean = players.select_dtypes(np.number).agg(['min', 'max', 'mean'])
print(min_max_mean)

# 6. 'market_value'와 'fpl_value' 열에 대해 0.91을 곱한 값 변환
transformed_values = players.loc[:, ['market_value', 'fpl_value']].transform(lambda x: x * 0.91)
print(transformed_values)

# 7. 'market_value'와 'fpl_value' 열에 대해 0.91을 곱한 값 계산
scaled_values = players.loc[:, ['market_value', 'fpl_value']] * 0.91
print(scaled_values)
```

```python
import numpy as np
import pandas as pd

# 예시 데이터 생성
data = {
    'age': [25.5, 30.2, 22.8, 29.9, 40.4],
    'height': [1.75, 1.80, 1.65, 1.72, 1.80],
    'club': ['ClubA', 'ClubB', 'ClubA', 'ClubC', 'ClubB']
}

players = pd.DataFrame(data)

# round_floats 함수 정의
def round_floats(x):
    if x.dtype == np.float64:
        return round(x)  # float64 타입일 경우 반올림
    return x  # 그 외의 경우 그대로 반환

# 1. apply를 사용하여 모든 열에 대해 round_floats 적용
rounded_players = players.apply(round_floats)

# 2. float64 타입의 열을 선택하여 head 출력
float64_columns_head = players.select_dtypes(np.float64).head()

# 3. float64 타입 열에만 apply로 round_floats 적용 후 head 출력
rounded_float64_columns_head = players.select_dtypes(np.float64).apply(round_floats).head()

# 결과 확인
print("Original DataFrame with rounded values:\n", rounded_players)
print("\nFirst 5 rows of float64 columns:\n", float64_columns_head)
print("\nFirst 5 rows after rounding float64 columns:\n", rounded_float64_columns_head)
```
