```python
import pandas as pd

# 리스트 정의
names = ['Olga', 'Andrew', 'Brian', 'Telulah', 'Nicole', 'Tilda']
ages = [29, 21, 45, 23, 39, 46]
married = [False, True, True, True, False, True]

# 시리즈 생성
ser = pd.Series(names, name='name')
print("▶ 이름 시리즈:")
print(ser)
```

```python

# 데이터프레임 생성
df = pd.DataFrame({
    'name': names,
    'age': ages,
    'married': married
})
```

```python
df
```

```python
# 특정 요소 접근
print("\n▶ ser에서 3번째 요소:", ser.iloc[2])      # Brian
print("▶ df에서 3번째 행, 첫 번째 열:", df.iloc[2, 0])  # Brian

# 시리즈와 데이터프레임의 크기 확인: 몇개 관측치가 있고 몇개 변수가 있는 지 확인
print("\n▶ ser의 shape:", ser.shape)
print("▶ df의 shape:", df.shape)

# 데이터 타입 확인
print("\n▶ df의 데이터 타입:")
print(df.dtypes) #info중에 정보 추출, 각 열별로 데이터 타입이 어던지

# 데이터프레임 정보 요약 출력
print("\n▶ df.info():")
df.info()

print("\n▶ df.info(verbose=True):")
df.info(verbose=True)
```

```python
# 데이터프레임 미리 보기
nutrition.head()

# 1. 특정 음식('Eggplant, raw')의 전체 영양 정보 (행 전체 반환)
nutrition.loc['Eggplant, raw']


# 2. 특정 음식의 특정 열 값 접근 (체인 인덱싱: 비추천)
nutrition.loc['Eggplant, raw']['calories']


# 3. 특정 음식의 특정 열 값 접근 (추천 방식)
nutrition.loc['Eggplant, raw', 'calories']


# 4. 음식 범위 & 영양소 범위 선택 (슬라이싱)
nutrition.loc['Eggplant, raw':'Sherbet, orange', 'calories':'cholesterol']


# 5. 특정 음식들 & 특정 영양소들 선택 (리스트 지정)
nutrition.loc[
    ['Raspberries, raw', 'Blackberries, raw'],
    ['protein', 'vitamin_b6', 'water']
]
```

```python
# 1. 4번째 행 전체
print("▶ nutrition.iloc[3]:")
print(nutrition.iloc[3])

# 2. 4번째 행 전체 (슬라이싱 사용)
print("\n▶ nutrition.iloc[3, :]:")
print(nutrition.iloc[3, :])



# 3. 5, 7, 10번째 행 전체
print("\n▶ nutrition.iloc[[4, 6, 9], :]:")
print(nutrition.iloc[[4, 6, 9], :])

# 4. 위와 동일 (모든 열은 기본 생략 가능)
print("\n▶ nutrition.iloc[[4, 6, 9]]:")
print(nutrition.iloc[[4, 6, 9]])


# 5. 5, 7, 10번째 행의 3번째 열만
print("\n▶ nutrition.iloc[[4, 6, 9], 2]:")
print(nutrition.iloc[[4, 6, 9], 2])


# 6. 5, 7, 10번째 행의 3~5번째 열
print("\n▶ nutrition.iloc[[4, 6, 9], 2:5]:")
print(nutrition.iloc[[4, 6, 9], 2:5])

```

```python
# 7. 짝수 인덱스 행과 짝수 인덱스 열만 필터링
new_nutr = nutrition.iloc[
    [True if i % 2 == 0 else False for i in range(8789)],
    [True if i % 2 == 0 else False for i in range(75)]
]
print("\n▶ 짝수 인덱스/열만 필터링한 new_nutr.shape:", new_nutr.shape)

```

```python
# ---------------------- LOC, AT, IAT ----------------------

# 8. 'Nuts, pecans' 행의 'calories' 값
print("\n▶ nutrition.loc['Nuts, pecans', 'calories']:")
print(nutrition.loc['Nuts, pecans', 'calories'])

# 9. 같은 위치의 값 (iloc 사용: 두 번째 행, 두 번째 열)
print("\n▶ nutrition.iloc[1, 1]:")
print(nutrition.iloc[1, 1])

# 10. .at을 이용한 접근 (빠름)
print("\n▶ nutrition.at['Nuts, pecans', 'calories']:")
print(nutrition.at['Nuts, pecans', 'calories'])

# 11. .iat을 이용한 위치 기반 접근
print("\n▶ nutrition.iat[1, 1]:")
print(nutrition.iat[1, 1])

# ---------------------- 성능 비교 ----------------------

print("\n▶ %timeit 비교 (loc vs at):")
# IPython 환경에서만 동작하는 %timeit을 대체하여 time 모듈로 비교
import time

start = time.time()
for _ in range(10000):
    nutrition.loc['Nuts, pecans', 'calories']
end = time.time()
print(f"loc 10000회: {end - start:.4f}초")

start = time.time()
for _ in range(10000):
    nutrition.at['Nuts, pecans', 'calories']
end = time.time()
print(f"at 10000회: {end - start:.4f}초")

# ---------------------- 컬럼 인덱스 번호 찾기 ----------------------

# 12. 'vitamin_k' 컬럼의 위치(숫자 인덱스) 찾기
print("\n▶ 'vitamin_k' 컬럼의 위치:")
print(nutrition.columns.get_loc('vitamin_k'))
```
