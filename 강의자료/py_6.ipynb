```python
import re

hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    # 이메일 패턴에 맞는 부분을 모두 찾아서 리스트로 반환
    x = re.findall('[a-zA-Z0-9]\S+@\S+[a-zA-Z]', line)
    if len(x) > 0:
        print(x)
```

```python
emails = set()
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    x = re.findall(r'[a-zA-Z0-9]\S+@\S+[a-zA-Z]', line)
    for mail in x:
        emails.add(mail)

for mail in sorted(emails):
    print(mail)
```

```python
import re

hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    # X-... 으로 시작하는 줄에서 콜론(:) 뒤에 오는 숫자(소수 포함)를 추출
    x = re.findall(r'^X\S*: ([0-9.]+)', line)
    if len(x) > 0:
        print(x)
```

```python
import re

hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    # "Details: rev=" 으로 시작하는 줄에서 숫자(점 포함)를 추출
    x = re.findall('^Details:rev', line)
    if len(x) > 0:
        print(x)
```

```python
import re

emails = [
    "nick@example.com",
    "nora@hanmail.net",
    "steve@gmail.com",
    "nancy.kim@naver.com"
]

pattern = re.compile(r'^n\S*@([A-Za-z0-9.-]+\.[A-Za-z]+)')

for mail in emails:
    m = pattern.match(mail)
    if m:
        print("이메일:", mail, "→ 도메인:", m.group(1))
```

```python
import re
p = re.compile('[a-z]+')   # 소문자 알파벳(a~z)이 1개 이상 연속되는 패턴

m = p.match("python")      # 문자열의 시작부터 이 패턴이 매칭되는지 확인
print(m)


m = p.match("3python")
print(m)   # None (시작이 숫자라 매칭 실패)

m = p.search("3python")
print(m)   # <re.Match object; span=(1, 7), match='python'>
```

```python
import xml.etree.ElementTree as ET
import pandas as pd

xml_data = """<data>
<country name="Liechtenstein">
  <rank>1</rank>
  <year>2008</year>
  <gdppc>141100</gdppc>
  <neighbor name="Austria" direction="E"/>
  <neighbor name="Switzerland" direction="W"/>
</country>
<country name="Singapore">
  <rank>4</rank>
  <year>2011</year>
  <gdppc>59900</gdppc>
  <neighbor name="Malaysia" direction="N"/>
</country>
<country name="Panama">
  <rank>68</rank>
  <year>2011</year>
  <gdppc>13600</gdppc>
  <neighbor name="Costa Rica" direction="W"/>
  <neighbor name="Colombia" direction="E"/>
</country>
</data>"""

# XML 파싱
root = ET.fromstring(xml_data)

rows = []
for c in root.findall("country"):
    row = {
        "country": c.attrib["name"],
        "rank": c.find("rank").text,
        "year": c.find("year").text,
        "gdppc": c.find("gdppc").text,
        # neighbor는 여러 개일 수 있으므로 join
        "neighbors": ", ".join([n.attrib["name"] for n in c.findall("neighbor")])
    }
    rows.append(row)

df = pd.DataFrame(rows)
print(df)
```

```python
import xml.etree.ElementTree as ET

# XML 파일 파싱
tree = ET.parse('country_data.xml')
root = tree.getroot()

print(root.tag)       # 최상위 태그 이름 출력 (예: "data")
print(root.attrib)    # 최상위 태그의 속성 딕셔너리 출력 (여기서는 {})

# root의 바로 아래 자식 노드 출력
for child in root:
    print(child.tag, child.attrib)

# root[0] : 첫 번째 <country>
# root[0][1] : 그 안의 두 번째 자식 태그 (<year>)
print(root[0][1].text)  # <year> 태그 안의 값 출력
```

```python
for neighbor in root.iter('neighbor'):
    print(neighbor.attrib)
```
