```python
import chardet
import pandas as pd

# 파일 경로 지정
file_path = 'BL-Flickr-Images-Book.csv'

# 파일을 바이너리 모드로 열어서 인코딩을 감지합니다.
with open(file_path, 'rb') as f:
    result = chardet.detect(f.read())

# 감지된 인코딩을 확인하고, 해당 인코딩으로 파일을 읽습니다.
encoding = result['encoding']
print(f"Detected encoding: {encoding}")

# pandas를 사용하여 해당 인코딩으로 CSV 파일 읽기
df = pd.read_csv(file_path, encoding=encoding)

# 데이터의 첫 5행을 출력하여 확인
df.head()
```

```python
# 불필요한 열 목록
to_drop = ['Edition Statement',        # Edition Statement 열
           'Corporate Author',         # Corporate Author 열
           'Corporate Contributors',   # Corporate Contributors 열
           'Former owner',             # Former owner 열
           'Engraver',                 # Engraver 열
           'Contributors',             # Contributors 열
           'Issuance type',            # Issuance type 열
           'Shelfmarks']               # Shelfmarks 열

# 'to_drop'에 나열된 열을 삭제합니다.
df.drop(to_drop, inplace=True, axis=1)  # axis=1은 열을 기준으로 삭제하는 것

# 'Identifier' 열을 인덱스로 설정합니다.
df.set_index('Identifier', inplace=True)

# 데이터프레임의 첫 5행을 출력하여 결과를 확인합니다.
df.head()
```

```python
# 첫번째  Cleaning
# 불필요한 문자 목록
unwanted_characters = ['[', ',', '-']  # [ , , - 문자를 제거할 예정

# 'Date of Publication' 열을 정리하는 함수 정의
def clean_dates(item):
    # 'Date of Publication' 값을 문자열로 변환 (nan 처리)
    dop = str(item.loc['Date of Publication'])

    # 'Date of Publication' 값이 'nan' 이거나, '['로 시작하면 NaN을 반환
    if dop == 'nan' or dop[0] == '[':
        return np.nan

    # 'Date of Publication'에 불필요한 문자가 포함되어 있으면 이를 제거
    for character in unwanted_characters:
        if character in dop:
            # 문자가 있는 첫 번째 위치에서 잘라냄
            character_index = dop.find(character)
            dop = dop[:character_index]

    # 정리된 'Date of Publication' 값을 반환
    return dop

# 'Date of Publication' 열에 대해 clean_dates 함수 적용
df['Date of Publication'] = df.apply(clean_dates, axis=1)

df.head()

```

```python
def clean_title(title):
    # title이 NaN이거나 float인 경우 'NaN' 반환
    if pd.isna(title) or not isinstance(title, str):
        return 'NaN'

    # 제목이 '['로 시작하면 ']'까지의 문자열을 추출하여 제목을 수정
    if title[0] == '[':
        title = title[1: title.find(']')]

    # 'by' 또는 'By'가 제목에 포함되면 그 부분까지 잘라냄
    if 'by' in title:
        title = title[:title.find('by')]
    elif 'By' in title:
        title = title[:title.find('By')]

    # 제목에 '['가 포함되어 있으면 그 부분까지 잘라냄
    if '[' in title:
        title = title[:title.find('[')]

    # 제목의 마지막 두 글자를 잘라냄 (예: 불필요한 공백 또는 문자)
    title = title[:-2]

    # 제목을 공백을 기준으로 분리한 후 각 단어의 첫 글자를 대문자로 변환
    title = list(map(str.capitalize, title.split()))

    # 각 단어를 공백으로 이어붙여서 최종 제목 반환
    return ' '.join(title)

# 'Title' 열에 대해 clean_title 함수 적용
df['Title'] = df['Title'].apply(clean_title)

# 데이터프레임의 첫 5행을 출력하여 결과 확인
df.head()
```

```python
# 'Place of Publication' 열을 가져옵니다.
pub = df['Place of Publication']

# 'London', 'Oxford', 'Newcastle upon Tyne'을 특정 형식으로 변경하고,
# 나머지 값은 그대로 유지하는 작업을 수행
df['Place of Publication'] = np.where(
    pub.str.contains('London'), 'London',  # 'London'이 포함된 경우 'London'으로 변경
    np.where(
        pub.str.contains('Oxford'), 'Oxford',  # 'Oxford'가 포함된 경우 'Oxford'로 변경
        np.where(
            pub.eq('Newcastle upon Tyne'), 'Newcastle-upon-Tyne',  # 정확히 'Newcastle upon Tyne'인 경우 'Newcastle-upon-Tyne'으로 변경
            df['Place of Publication']  # 그 외의 경우 원래 값을 그대로 유지
        )
    )
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import datasets

# Iris 데이터셋 로드
iris = datasets.load_iris()

# 특성 데이터프레임 생성
dataset = pd.DataFrame(iris.data)
dataset.columns = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width']

# 레이블 (labels) 생성
labels = pd.DataFrame(iris.target)
labels.columns = ['labels']

# 특성 데이터와 레이블 데이터 결합
data = pd.concat([dataset, labels], axis=1)

# 2x2 레이아웃으로 서브플롯 만들기
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 각 서브플롯에 히스토그램 그리기
sns.histplot(data=data, x='Sepal length', hue='labels', multiple='stack', palette='Set1', kde=False, ax=axes[0, 0])
axes[0, 0].set_title('Histogram of Sepal Length')

sns.histplot(data=data, x='Sepal width', hue='labels', multiple='stack', palette='Set1', kde=False, ax=axes[0, 1])
axes[0, 1].set_title('Histogram of Sepal Width')

sns.histplot(data=data, x='Petal length', hue='labels', multiple='stack', palette='Set1', kde=False, ax=axes[1, 0])
axes[1, 0].set_title('Histogram of Petal Length')

sns.histplot(data=data, x='Petal width', hue='labels', multiple='stack', palette='Set1', kde=False, ax=axes[1, 1])
axes[1, 1].set_title('Histogram of Petal Width')

# 그래프 간격 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```
