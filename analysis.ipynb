```python
# 월별 GitHub 레포지토리 생성 건수
github_monthly = github_df.groupby(['year_month', 'period']).agg({
    'id': 'count',
    'stars': 'sum',
    'forks': 'sum',
    'open_issues': 'sum'
}).reset_index()

github_monthly.columns = ['year_month', 'period', 'repo_count', 'total_stars', 'total_forks', 'total_issues']

print("GitHub 월별 통계:")
display(github_monthly.head(10))
```

## 7. 회귀분석

```python
# Period 더미 변수 추가
merged_monthly_with_period = pd.merge(
    merged_monthly,
    github_monthly[['year_month', 'period']].drop_duplicates(),
    on='year_month',
    how='left'
)

# Period를 0/1로 변환 (Before=0, After=1)
merged_monthly_with_period['period_dummy'] = (merged_monthly_with_period['period'] == 'After').astype(int)

# AI 관련 프로젝트 비율 계산 (월별)
github_ai_ratio = github_df.groupby('year_month').apply(
    lambda x: (x['ai_field'].notna().sum() / len(x)) if len(x) > 0 else 0
).reset_index(name='ai_ratio')

merged_monthly_with_period = pd.merge(
    merged_monthly_with_period,
    github_ai_ratio,
    on='year_month',
    how='left'
)

print("회귀분석용 데이터:")
display(merged_monthly_with_period.head())
```

```python
# 다중회귀분석 모델
# 종속변수: Stack Overflow 질문 수
# 독립변수: GitHub 활동 지표들 + Period + AI 비율

X = merged_monthly_with_period[['repo_count', 'total_issues', 'total_forks', 'period_dummy', 'ai_ratio']]
y = merged_monthly_with_period['question_count']

# 상수항 추가
X = sm.add_constant(X)

# 회귀모형 적합
model = sm.OLS(y, X).fit()

print("=" * 80)
print("다중회귀분석 결과")
print("=" * 80)
print(model.summary())
```

```python
# 회귀계수 해석
print("\n" + "=" * 80)
print("회귀계수 해석")
print("=" * 80)

coefficients = model.params
pvalues = model.pvalues

print(f"\n1. 상수항: {coefficients['const']:.2f}")
print(f"   - p-value: {pvalues['const']:.4f}")

print(f"\n2. GitHub 레포지토리 수: {coefficients['repo_count']:.4f}")
print(f"   - 해석: 레포지토리가 1개 증가하면 SO 질문은 {coefficients['repo_count']:.4f}개 {'증가' if coefficients['repo_count'] > 0 else '감소'}")
print(f"   - p-value: {pvalues['repo_count']:.4f} ({'유의함' if pvalues['repo_count'] < 0.05 else '유의하지 않음'})")

print(f"\n3. GitHub Issues: {coefficients['total_issues']:.4f}")
print(f"   - 해석: Issues가 1개 증가하면 SO 질문은 {coefficients['total_issues']:.4f}개 {'증가' if coefficients['total_issues'] > 0 else '감소'}")
print(f"   - p-value: {pvalues['total_issues']:.4f} ({'유의함' if pvalues['total_issues'] < 0.05 else '유의하지 않음'})")

print(f"\n4. GitHub Forks: {coefficients['total_forks']:.4f}")
print(f"   - 해석: Forks가 1개 증가하면 SO 질문은 {coefficients['total_forks']:.4f}개 {'증가' if coefficients['total_forks'] > 0 else '감소'}")
print(f"   - p-value: {pvalues['total_forks']:.4f} ({'유의함' if pvalues['total_forks'] < 0.05 else '유의하지 않음'})")

print(f"\n5. Period (ChatGPT 출시 후): {coefficients['period_dummy']:.2f}")
print(f"   - 해석: ChatGPT 출시 후 SO 질문은 평균 {abs(coefficients['period_dummy']):.2f}개 {'증가' if coefficients['period_dummy'] > 0 else '감소'}")
print(f"   - p-value: {pvalues['period_dummy']:.4f} ({'유의함' if pvalues['period_dummy'] < 0.05 else '유의하지 않음'})")

print(f"\n6. AI 프로젝트 비율: {coefficients['ai_ratio']:.2f}")
print(f"   - 해석: AI 비율이 1% 증가하면 SO 질문은 {coefficients['ai_ratio']:.2f}개 {'증가' if coefficients['ai_ratio'] > 0 else '감소'}")
print(f"   - p-value: {pvalues['ai_ratio']:.4f} ({'유의함' if pvalues['ai_ratio'] < 0.05 else '유의하지 않음'})")

print(f"\n모델 적합도:")
print(f"  R-squared: {model.rsquared:.4f}")
print(f"  Adjusted R-squared: {model.rsquared_adj:.4f}")
print(f"  F-statistic: {model.fvalue:.2f}")
print(f"  Prob (F-statistic): {model.f_pvalue:.4e}")
```

```python
# 그래프 7: 회귀분석 시각화 - 실제값 vs 예측값
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# 1) 실제값 vs 예측값
predicted = model.predict(X)
axes[0].scatter(y, predicted, alpha=0.6)
axes[0].plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
axes[0].set_xlabel('실제 Stack Overflow 질문 수', fontsize=11)
axes[0].set_ylabel('예측 Stack Overflow 질문 수', fontsize=11)
axes[0].set_title('실제값 vs 예측값', fontsize=12, fontweight='bold')
axes[0].grid(True, alpha=0.3)

# 2) 잔차 플롯
residuals = y - predicted
axes[1].scatter(predicted, residuals, alpha=0.6)
axes[1].axhline(y=0, color='r', linestyle='--', lw=2)
axes[1].set_xlabel('예측값', fontsize=11)
axes[1].set_ylabel('잔차', fontsize=11)
axes[1].set_title('잔차 플롯', fontsize=12, fontweight='bold')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('images/regression_diagnostics.png', dpi=300, bbox_inches='tight')
plt.show()

print("그래프 저장 완료: images/regression_diagnostics.png")
```

```python
# 그래프 8: 회귀계수 시각화
fig, ax = plt.subplots(figsize=(10, 6))

# 상수항 제외한 계수들
coef_names = ['GitHub\n레포지토리', 'GitHub\nIssues', 'GitHub\nForks', 'Period\n(After)', 'AI 비율']
coef_values = coefficients[1:].values
coef_colors = ['green' if v > 0 else 'red' for v in coef_values]

bars = ax.barh(coef_names, coef_values, color=coef_colors, alpha=0.7)
ax.axvline(x=0, color='black', linestyle='-', linewidth=0.8)
ax.set_xlabel('회귀계수', fontsize=12)
ax.set_title('Stack Overflow 질문 수에 미치는 영향 (회귀계수)', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3, axis='x')

# 유의성 표시
for i, (bar, pval) in enumerate(zip(bars, pvalues[1:])):
    if pval < 0.001:
        ax.text(bar.get_width(), bar.get_y() + bar.get_height()/2, ' ***', 
               va='center', fontsize=12, fontweight='bold')
    elif pval < 0.01:
        ax.text(bar.get_width(), bar.get_y() + bar.get_height()/2, ' **', 
               va='center', fontsize=12, fontweight='bold')
    elif pval < 0.05:
        ax.text(bar.get_width(), bar.get_y() + bar.get_height()/2, ' *', 
               va='center', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.savefig('images/regression_coefficients.png', dpi=300, bbox_inches='tight')
plt.show()

print("그래프 저장 완료: images/regression_coefficients.png")
print("\n유의성 표시: *** p<0.001, ** p<0.01, * p<0.05")
```
